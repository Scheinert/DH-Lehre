```python
# Excel-Datei laden
import pandas as pd
import numpy as np
import networkx as nx
import plotly.graph_objects as go

file_path = 'Liste_Version_2.xlsx'  # Ersetze dies durch den tatsächlichen Pfad
df = pd.read_excel(file_path)
df.columns = df.columns.str.strip()

# Überprüfe, ob die erforderlichen Spalten existieren
required_columns = ['Bereich', 'Fachbereich', 'Name_des_Tools']
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"Die erforderlichen Spalten {required_columns} sind nicht in der Datei vorhanden.")

# Daten bereinigen und relevante Spalten auswählen
df.columns = df.columns.str.strip()
edges = df[required_columns].dropna()

# Graphen erstellen
G = nx.Graph()

# Kanten zwischen Bereich, Fachbereich und Name des Tools hinzufügen
for _, row in edges.iterrows():
    G.add_edge(row['Bereich'], row['Fachbereich'])
    G.add_edge(row['Fachbereich'], row['Name_des_Tools'])

# 3D Layout für den Graphen
pos = nx.spring_layout(G, dim=3)

# Knotenpositionen extrahieren
x_vals = [pos[node][0] for node in G.nodes]
y_vals = [pos[node][1] for node in G.nodes]
z_vals = [pos[node][2] for node in G.nodes]

# Sicherstellen, dass der Punkt "DH" existiert
if 'DH' not in G.nodes:
    raise ValueError('Der Knoten "DH" existiert nicht im Graphen.')

# Position von "DH"
dh_pos = pos['DH']
dh_x, dh_y, dh_z = dh_pos[0], dh_pos[1], dh_pos[2]

# Plotly-Spur für Knoten
node_trace = go.Scatter3d(
    x=x_vals, y=y_vals, z=z_vals,
    mode='markers+text',
    text=list(G.nodes),
    textposition='top center',
    marker=dict(size=6, color='lightblue', opacity=0.8)
)

# Plotly-Spur für Kanten
edge_trace = []
for edge in G.edges():
    x0, y0, z0 = pos[edge[0]]
    x1, y1, z1 = pos[edge[1]]
    edge_trace.append(go.Scatter3d(
        x=[x0, x1, None], y=[y0, y1, None], z=[z0, z1, None],
        mode='lines',
        line=dict(color='gray', width=1),
        hoverinfo='none'
    ))

# Traces kombinieren
fig = go.Figure(data=[*edge_trace, node_trace])

# Layout anpassen
fig.update_layout(
    title='Tools der Digital Humanities',
    title_x=0.5,
    showlegend=False,
    scene=dict(
        xaxis=dict(
            showbackground=False, showticklabels=False, title=None, # X-Achse deaktiviert
            zeroline=False, showgrid=False, showline=False, ticks=''
        ),
        yaxis=dict(
            showbackground=False, showticklabels=False, title=None, # Y-Achse deaktiviert
            zeroline=False, showgrid=False, showline=False, ticks=''
        ),
        zaxis=dict(
            showbackground=False, showticklabels=False, title=None, # Z-Achse deaktiviert
            zeroline=False, showgrid=False, showline=False, ticks=''
        ),
        camera=dict(
            eye=dict(x=1.25, y=1.25, z=1.25)  # Initiale Kameraansicht
        )
    )
)

# Button für die Rotation hinzufügen
buttons = []  # Initialisiere die Liste der Buttons

# Animation für Rotation um "DH"
frames = []
angle_step = 1  # Schrittweite für jeden Frame
for angle in range(0, 360, angle_step):
    rad = np.radians(angle)  # Umwandlung von Grad in Radianten
    # Neue Kamera-Positionen berechnen (um "DH" rotieren)
    eye_x = dh_x + 2 * np.cos(rad)
    eye_y = dh_y + 2 * np.sin(rad)
    eye_z = dh_z + 0.5  # Z-Achse bleibt konstant
    
    frames.append(dict(
        layout=dict(
            scene_camera=dict(
                eye=dict(x=eye_x, y=eye_y, z=eye_z)
            )
        )
    ))

# Hinzufügen der Frames für die Animation
fig.frames = frames

# Animation konfigurieren
fig.update_layout(
    updatemenus=[dict(
        type="buttons",
        showactive=False,
        buttons=[dict(label="Play",
                      method="animate",
                      args=[None, dict(frame=dict(duration=50, redraw=True),
                                       fromcurrent=True, mode='immediate')])]
    )]
)


    # Füge die Frames zur Figur hinzu
    fig.frames = frames

# Füge die Buttons dem Layout hinzu
fig.update_layout(
    updatemenus=[dict(
        type="buttons",
        showactive=True,
        buttons=buttons,
        x=0.01,
        y=0.5,
        xanchor='left',
        yanchor='top'
    )]
)

# Graph anzeigen und speichern
fig.show()
fig.write_html("fig4.html")
```

```python

import pandas as pd
import networkx as nx
import plotly.graph_objects as go
import numpy as np
import plotly.io as pio
pio.renderers.default = 'browser'

# Excel-Datei laden (Beispiel für Datenpfad; bitte anpassen)
file_path = 'Liste_Version_2.xlsx'  # Ersetze dies durch den tatsächlichen Pfad
df = pd.read_excel(file_path)

# Überprüfe, ob die erforderlichen Spalten existieren
required_columns = ['Bereich', 'Fachbereich', 'Name_des_Tools']
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"Die erforderlichen Spalten {required_columns} sind nicht in der Datei vorhanden.")

# Daten bereinigen und relevante Spalten auswählen
df.columns = df.columns.str.strip()
edges = df[required_columns].dropna()

# Graphen erstellen
G = nx.Graph()

# Kanten zwischen Bereich, Fachbereich und Name des Tools hinzufügen
for _, row in edges.iterrows():
    G.add_edge(row['Bereich'], row['Fachbereich'])
    G.add_edge(row['Fachbereich'], row['Name_des_Tools'])

# Positionen für Knoten im Layout berechnen
pos = nx.spring_layout(G)

# Kanten für den Plot erfassen
edge_trace = go.Scatter(
    x=[],
    y=[],
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines')

for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_trace['x'] += [x0, x1, None]
    edge_trace['y'] += [y0, y1, None]

# Knoten für den Plot erfassen
node_trace = go.Scatter(
    x=[],
    y=[],
    text=[],
    mode='markers+text',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        size=10,
        color=[],
        colorbar=dict(
            thickness=15,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        )
    )
)

# Fülle die Knoten-Daten
for node in G.nodes():
    x, y = pos[node]
    node_trace['x'] += [x]
    node_trace['y'] += [y]
    node_trace['text'] += [node]

# Layout definieren
layout = go.Layout(
    title='<br>Netzwerkgraph',
    titlefont_size=16,
    showlegend=False,
    hovermode='closest',
    margin=dict(b=0, l=0, r=0, t=40),
    xaxis=dict(showgrid=False, zeroline=False),
    yaxis=dict(showgrid=False, zeroline=False)
)

# Figur erstellen
fig = go.Figure(data=[edge_trace, node_trace], layout=layout)

# Optional: Animation für Rotation hinzufügen
frames = []
angle_step = 1  # Schrittweite für jeden Frame
for angle in range(0, 360, angle_step):
    rad = np.radians(angle)
    eye_x = 2 * np.cos(rad)
    eye_y = 2 * np.sin(rad)
    eye_z = 0.5
    frames.append(dict(
        layout=dict(
            scene_camera=dict(
                eye=dict(x=eye_x, y=eye_y, z=eye_z)
            )
        )
    ))

# Füge die Frames einmalig hinzu
fig.frames = frames

# Animation konfigurieren
fig.update_layout(
    updatemenus=[dict(
        type="buttons",
        showactive=False,
        buttons=[dict(label="Play",
                      method="animate",
                      args=[None, dict(frame=dict(duration=50, redraw=True),
                                       fromcurrent=True, mode='immediate')])]
    )]
)

# Graph anzeigen und speichern
fig.show()
```
